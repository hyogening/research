```python
import numpy as np
import pandas as pd
 
df = pd.read_csv('HVIP_sorted.csv')
df.head()
```

```python
df['vocationaluse'] = df['vocationaluse'].fillna('※ Infrastructure')
df
```

```python
df1 = df.replace({'vocationaluse': {'Refuse':'Misc / other', 'Truck - Medium Duty': 'Misc / other', 'Truck - Pickup': 'Misc / other', 'Utility Truck': 'Misc / other'}})
df2 = df1.groupby(['county', 'vocationaluse', 'payment_year'])[['amount']].sum()
df3 = df2.unstack(level=2)
```

```python
df3.to_csv("HVIP Funded Vehicles.csv")
```
